```python
class Node:
    def __init__(self,data):
        self.prev=None
        self.data=data
        self.next=None
```

```python
class SinglyLinkedList:
    def __init__(self,*data):
        self.head=None
        if len(data)==0:
                print("Linked list is empty!")
        elif len(data)==1:
            self.head=Node(data)
        else:
            p=None
            for i in range(len(data)):
                if i==0:
                    self.head=Node(data[i])
                    p=self.head
                else:
                    p.next=Node(data[i])
                    p=p.next
                    
    def addFirstNode(self,data):
        m=self.head
        self.head=Node(data)
        self.head.next=m
#         return self.head
#         return self.head.data

    def addLastNode(self,data):
        m=self.head
        while m.next!=None:
            m=m.next
        m.next=Node(data)
#         return self.head
#         return self.head.data

    def insertNode(self,index,data):
        m=self.head
        idx=-1
        if index<0 or index>self.countNode():
            print("Invalid Index")
        else:
            while m!=None:
                idx+=1
                if index==0:
                    self.addFirstNode(data)
                    break
                elif index==self.countNode()-1:
                    self.addLastNode(data)
                    break
                else:
                    if idx==index-1:
                        n=m.next
                        m.next=Node(data)
                        m.next.next=n
                        break 
                m=m.next
                
    def len(self):
        return self.countNode()-1

    def indexOfNode(self,data):
        m=self.head
        index=-1
        valueError=True
        while m!=None:
            index+=1
            if m.data==data:
                valueError=False
                break
            m=m.next
        if valueError:
            index=-1
        return index

    def countNode(self):
        m=self.head
        count=0
        while m!=None:
            count+=1
            m=m.next
        return count
    
    def reversePrint(self):
        m=self.head
        n=SinglyLinkedList()
        while m!=None:
            n.addFirstNode(m.data)
            m=m.next
        n.printList()
            
    
    def printList(self):
        m=self.head
        print(">>>",end="")
        while m!=None:
            print(m.data,end=" ")
            m=m.next
        print()
```
